MaLSTM
- LSTM층을 적용해 각 문장의 의미벡터를 뽑는다.
- 의미 벡터는 각 LSTM의 마지막 스텝의 값이 은닉 상태 벡터로 사용된다.
- 이 값은 문장의 모든 단어에 대한 정보가 반영된 값으로 전체 문장을 대표한는 벡터가 된다.
- 이렇게 뽑은 두 벡터에 대해 맨하탄 거리를 계산해서 두 문장 사이의 유사도를 측정한다.
- 이 유사도를 실제 라벨과 비교해서 학습하는 방식이다.

```python
import numpy as no
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Input, Embedding, LSTM, Layer
from tensorflow.keras.layers import Dense, Dropout, Concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import regularizers
import tensorflow.keras.backend as K
import matplotlib.pyplot as plt
```

```python
DATA_IN_PATH = 'C:/inkyun/실습파일과 교재/5.자연어처리(실습파일)/dataset/'
train_q1_data_file = '5-2.train_q1.npy'
train_q2_data_file = '5-2.train_q2.npy'
train_label_data_file = '5-2.train_label.npy'

q1_data = np.load(open(DATA_IN_PATH + train_q1_data_file, 'rb'))
q2_data = np.load(open(DATA_IN_PATH + train_q2_data_file, 'rb'))
labels = np.load(open(DATA_IN_PATH + train_label_data_file, 'rb'))
```

```python
trainQ1, testQ1, trainQ2, testQ2, trainY, testY = train_test_split(q1_data, q2_data, labels, test_size=0.2)
```

```python
vocab_size = np.max([q1_data, q2_data]) + 1
emb_size = 16
hidden_dim = 10
feature_dim = 10
regularizer = 0.005
```

```python
K.clear_session()
inputQ1 = Input(batch_shape=(None, trainQ1.shape[1]))
inputQ2 = Input(batch_shape=(None, trainQ2.shape[1]))

embQ1 = Embedding(input_dim=vocab_size, output_dim = emb_size)(inputQ1)
embQ1 = Dropout(rate=0.2)(embQ1)
lstmQ1 = LSTM(hidden_dim)(embQ1)
lstmQ1 = Dense(feature_dim, activation = 'relu', kernel_regularizer=regularizers.l2(regularizer))(lstmQ1)
lstmQ1 = Dropout(rate=0.2)(lstmQ1)
```

```python
embQ2 = Embedding(input_dim=vocab_size, output_dim = emb_size)(inputQ2)
embQ2 = Dropout(rate=0.2)(embQ2)
lstmQ2 = LSTM(hidden_dim)(embQ2)
lstmQ2 = Dense(feature_dim, activation = 'relu', kernel_regularizer=regularizers.l2(regularizer))(lstmQ2)
lstmQ2 = Dropout(rate=0.2)(lstmQ2)
```

```python
mDist = K.exp(-K.sum(K.abs(lstmQ1 - lstmQ2), axis=1, keepdims = True))

model = Model([inputQ1, inputQ2], mDist)
model.compile(loss = 'mean_squared_error', optimizer = Adam(lr=0.0005))
model.summary()
```

```python
trainY = trainY.reshape(-1,1)
testY = testY.reshape(-1,1)
hist = model.fit([trainQ1, trainQ2], trainY, validation_data = ([testQ1, testQ2], testY), batch_size = 1000, epochs = 10)
```

```python
predicted = model.predict([testQ1, testQ2])
predY = np.where(predicted > 0.5, 1, 0)
accuracy = (testY == predY).mean()
print(accuracy * 100)
```

Embedding layer 1개로 구성해보기

```python
inputQ1 = Input(batch_shape=(None, trainQ1.shape[1]))
inputQ2 = Input(batch_shape=(None, trainQ2.shape[1]))

embQ = Embedding(input_dim=vocab_size, output_dim=emb_size)
embQ1 = embQ(inputQ1)
embQ2 = embQ(inputQ2)

lstmQ1 = LSTM(hidden_dim)(embQ1)
lstmQ1 = Dense(feature_dim, activation = 'relu', kernel_regularizer=regularizers.l2(regularizer))(lstmQ1)
lstmQ1 = Dropout(rate=0.2)(lstmQ1)

lstmQ2 = LSTM(hidden_dim)(embQ2)
lstmQ2 = Dense(feature_dim, activation = 'relu', kernel_regularizer=regularizers.l2(regularizer))(lstmQ2)
lstmQ2 = Dropout(rate=0.2)(lstmQ2)

mDist = K.exp(-K.sum(K.abs(lstmQ1 - lstmQ2), axis=1, keepdims = True))

model = Model([inputQ1, inputQ2], mDist)
model.compile(loss = 'mean_squared_error', optimizer = Adam(lr=0.0005))
# model.summary()
```

```python
trainY = trainY.reshape(-1,1)
testY = testY.reshape(-1,1)
hist = model.fit([trainQ1, trainQ2], trainY, validation_data = ([testQ1, testQ2], testY), batch_size = 1000, epochs = 10)
```

```python
predicted = model.predict([testQ1, testQ2])
predY = np.where(predicted > 0.5, 1, 0)
accuracy = (testY == predY).mean()
print(accuracy * 100)
```

```python
# Quora question pairs : 전처리
# -----------------------------
import pandas as pd
import numpy as np
import re
from tensorflow.python.keras.preprocessing.text import Tokenizer
from tensorflow.python.keras.preprocessing.sequence import pad_sequences
import pickle

# Quora question 데이터를 읽어온다.
# DATA_IN_PATH = './dataset/'
train_data = pd.read_csv('C:/inkyun/실습파일과 교재/5.자연어처리(실습파일)/dataset/'+ '5-1.questions.csv')
train_data.head()

# 중복된 페어와 중복되지 않은 페어로 분리한다.
train_pos_data = train_data.loc[train_data['is_duplicate'] == 1]
train_neg_data = train_data.loc[train_data['is_duplicate'] == 0]

# 중복되지 않은 페어가 많으므로 둘의 비율이 비슷하도록 양을 조절한다.
# dample_frac (%) 만큼 샘플링
sample_frac = len(train_pos_data) / len(train_neg_data)
print("before : %.2f" % sample_frac)

train_neg_data = train_neg_data.sample(frac = sample_frac)
sample_frac = len(train_pos_data) / len(train_neg_data)
print("after : %.2f" % sample_frac)

train_pos_data.head()
train_neg_data.head()

# 두 데이터를 다시 합친다
train_data = pd.concat([train_neg_data, train_pos_data])

# 간단히 전처리를 수행한다. FILTERS에 포함된 문자 제거, 소문자로 변환.
FILTERS = "([~.,!?\"':;)(])"
change_filter = re.compile(FILTERS)

questions1 = [str(s) for s in train_data['question1']]
questions2 = [str(s) for s in train_data['question2']]

filtered_questions1 = list()
filtered_questions2 = list()

for q in questions1:
     filtered_questions1.append(re.sub(change_filter, "", q).lower())
        
for q in questions2:
     filtered_questions2.append(re.sub(change_filter, "", q).lower())

# volcabulary를 구축하고 단어들을 워드 인덱스로 변환한다.
# 한 문장의 길이는 31개로 제한한다.
tokenizer = Tokenizer()
tokenizer.fit_on_texts(filtered_questions1 + filtered_questions2)
questions1_sequence = tokenizer.texts_to_sequences(filtered_questions1)
questions2_sequence = tokenizer.texts_to_sequences(filtered_questions2)

MAX_SEQUENCE_LENGTH = 31
q1_data = pad_sequences(questions1_sequence, maxlen=MAX_SEQUENCE_LENGTH, 
                        padding='post', truncating='post')
q2_data = pad_sequences(questions2_sequence, maxlen=MAX_SEQUENCE_LENGTH, 
                        padding='post', truncating='post')
labels = np.array(train_data['is_duplicate'], dtype=int)

# 전처리 결과를 저장해 둔다.
TRAIN_Q1_DATA = '5-2.train_q1.npy'
TRAIN_Q2_DATA = '5-2.train_q2.npy'
TRAIN_LABEL_DATA = '5-2.train_label.npy'

np.save(open(DATA_IN_PATH + TRAIN_Q1_DATA, 'wb'), q1_data)
np.save(open(DATA_IN_PATH + TRAIN_Q2_DATA , 'wb'), q2_data)
np.save(open(DATA_IN_PATH + TRAIN_LABEL_DATA , 'wb'), labels)

# 결과를 저장한다.
with open('C:/inkyun/실습파일과 교재/5.자연어처리(실습파일)/dataset/5-2.vocabulary.pickle', 'wb') as f:
    pickle.dump(tokenizer.word_index, f, pickle.HIGHEST_PROTOCOL)
```
